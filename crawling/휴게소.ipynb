### 노선별, 방향별 휴게소 편의시설 현황 (경부선)

```python
import requests
from urllib.request import urlopen
from bs4 import BeautifulSoup
import pandas as pd
```

```python
routeName = '%EA%B2%BD%EB%B6%80%EC%84%A0' # 경부선
service = '8500240533'
url =f"http://data.ex.co.kr/openapi/business/conveniServiceArea?key={service}&type=xml&numOfRows=50&pageNo=1&routeName={routeName}"
print(url)
```

#### 페이지 읽기

```python
xml = requests.get(url).text
soup = BeautifulSoup(xml, 'xml')
```

```python
area = soup.find_all('list')
len(area)
# area[0].find_all()
```

```python
# 휴게소 하나의 데이터를 딕셔너리에 추가 함수
def to_extract_row(area):
    data = {}
    for tag in area.find_all():
        data[tag.name] = tag.text if tag.text else None
    return data
```

```python
# 항목명이 있을 때 휴게소 하나의 데이터를 딕셔너리에 추가하는 함수
def to_extract_row_col(area, col_names=None):
    data = {}
    if col_names is None:
        to_extract_row(area)
    else:
        for col, tag in zip(col_names, area.find_all()):
            data[col] = tag.text if tag.text else None
    return data

col_names = ['브랜드', '편의시설', '방향', '경정비유무', '노선코드', '노선이름', '영업부대시설코드', '휴게소/주유소코드', '휴게소/주유소명', '휴게소주소', '전화번호', '화물휴게소유무']
```

```python
# 딕셔너리 호출 함수
def make_df(area):
    dataList = [to_extract_row(ar) for ar in area]
    return pd.DataFrame(dataList)
```

```python
make_df(area)
```

```python
area_df = make_df(area)
area_df.to_csv('crawl_data/휴게소_경부선.csv', index=False)
```
