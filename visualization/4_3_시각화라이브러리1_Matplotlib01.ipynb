```python
plt.figure(figsize=(3,3))
plt.plot([1,4,9,16])
plt.title('Line Plot')
plt.ylim(0,20)
plt.xticks([0,1,2,3], ['Jan', 'Feb', 'Mar', 'Apr'], rotation=45)
plt.yticks([0,5,10,20])
plt.xlabel('Season')
plt.ylabel('Y')
plt.show()
```

## 4. plot()

### plot() 함수

- 기본으로 선을 그리는 함수
- 데이터가 시간, 순서 등에 따라 변화를 보여주기 위해 사용


**형식** :  

matplotlib.pyplot.plot(*args, scalex=True, scaley=True, data=None, **kwargs)

: *args : x, y 축값을 리스트 형식으로 입력, y축값만 쓰는 경우 x축값은 자동 생성

https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html

```python
x = np.linspace(0, 2*np.pi, 50)
y = np.sin(x)
plt.figure(figsize=(3,3))
plt.plot(x, y)
plt.show()
```

#### 마커 선굵기, 색상 변경

- 마커 선굵기 : markeredgewidth (mew)
- 마커 내부색깔: markerfacecolor (mfc)

```python
x = list(range(7))
y = [1,4,5,8,9,5,3]
plt.figure(figsize=(5,3))
plt.plot(x, y, color='green', linestyle='dashed', marker='o',
        markersize=10, markerfacecolor='blue', markeredgecolor='red', markeredgewidth=2)
plt.show()
```

```python
# 스타일을(마커, 선) 약자로 표사
plt.figure(figsize=(5,3))
plt.plot(x, y, color='b', ls='-.', marker='o',
        ms=8, mfc='g', mec='r', mew=2)
plt.show()
```

#### 선의 색상과 스타일을 한번에 지정하기

```python
x2 = [i+2 for i in x]
plt.figure(figsize=(5,3))
plt.plot(x, y, 'g-.')
plt.plot(x2, y, 'r:')
plt.title('선색상과 선스타일을 한번에 지정')
plt.show()
```

#### 선의 색상과 스타일, 마커모양을 한번에 지정하기

```python
x2 = [i+2 for i in x]
plt.figure(figsize=(5,3))
plt.plot(x, y, 'g-.o')
plt.plot(x2, y, 'r:*')
plt.title('선색상과 선스타일, 마커모양을 한번에 지정')
plt.show()
```

## 5. 여러 개의 플롯을 동일한 그림영역 내에 그리기

- plot()을 선그래프 수만큼 지정하거나, 하나의 plot() 함수 내에 지정하여 호출

```python
x = np.arange(1,26)
y = np.arange(0, 5, 0.2)
```

```python
plt.figure(figsize=(4,3))
plt.title('multi-line plot')
plt.plot(x, y, 'r--', label='plot1')
plt.plot(x, 0.5*y**2, 'bs:', label='plot2')
plt.plot(x, 0.2*y**3, 'g^-', label='plot3')
plt.legend()
plt.show()
```

#### 위 그래프 코드를 plot() 하나로 한번에 표현하기

```python
plt.figure(figsize=(4,3))
plt.title('multi-line plot')
plt.plot(x, y, 'r--', x, 0.5*y**2, 'bs:', x, 0.2*y**3, 'g^-')
plt.show()
```

## 6.  하나의 그림 영역을 분할하여 여러 플롯 표시

### 1) plt.subplot() 함수

하나의 윈도우(figure)안에 여러 개의 플롯을 배열 형태로 표시

- 그리드 형태의 Axes 객체들 생성

- 형식 : subplot(인수1, 인수2, 인수3)
    - 인수1 과 인수2는 전체 그리드 행렬 모양 지시
    - 인수3 : 그래프 위치 번호


    - subplot(2,2,1) 가 원칙이나 줄여서 221로 쓸 수 있음
    - subplot(221) 2행 2열의 그리드에서 첫번째 위치
    - subplot(224) 2행 2열의 그리드에서 마지막 위치
    

- subplot()은 show()하기 전에 먼저 위치 설정해야 함

https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.subplot.html

 - tight_layout(pad=) : 플롯간 간격을 설정
    - pad = 간격값(실수)

https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.tight_layout.html

#### 2 x 2 형태의 네 개의 플롯

```python
np.random.seed(0)
plt.figure(figsize=(6,5))

# plt.subplot(2,2,1)
plt.subplot(221)
plt.plot(np.random.rand(5))
plt.title('axes1')

plt.subplot(222)
plt.plot(np.random.rand(5))
plt.title('axes2')

plt.subplot(223)
plt.plot(np.random.rand(5))
plt.title('axes3')

plt.subplot(224)
plt.plot(np.random.rand(5))
plt.title('axes4')

plt.tight_layout()
plt.show()
```

### 3) plt.subplot_mosaic() 함수

- 서로 다른 크기의 영역으로 구분하여 플롯 그리기
- 반환값은 {figure, axes_dict}, axes_dict = {'영역라벨':axes객체,...}

https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.subplot_mosaic.html

```python
fig, axes_dict = plt.subplot_mosaic([['A', 'right'], ['B', 'right']], 
                                    figsize=(4,3), layout='constrained')

for ax_name in axes_dict:
    axes_dict[ax_name].text(0.5, 0.5, ax_name, ha='center', va='center')
```

```python
fig = plt.figure(figsize=(4,3), layout='constrained', facecolor='lightskyblue')
fig.suptitle('Figure')
figL, figR = fig.subfigures(1,2)
figL.set_facecolor('thistle')
figL.suptitle('Left subfigure')
axL = figL.subplots(2,1, sharex=True)
axL[1].set_xlabel('x [m]')
figL.set_facecolor('paleturquoise')
figL.suptitle('Right subfigure')
axR = figR.subplots(1,2, sharey=True)
axR[0].set_title('Axes 1')

plt.show()
```

```python
x = [1,2,3,4,5]
y1 = [1,4,9,16,25]
y2 = [25,16,9,4,1]
y3 = [5,7,6,8,7]

fig, ax = plt.subplot_mosaic([['A'],['B'],['C']], figsize=(3,6))
# fig, ax = plt.subplot_mosaic(['A','B','C'], figsize=(3,6))

ax['A'].plot(x, y1)
ax['A'].set_title('Line plot1')
ax['B'].plot(x, y2)
ax['B'].set_title('Line plot1')
ax['C'].plot(x, y3)
ax['C'].set_title('Line plot1')

plt.tight_layout()
plt.show()
```

#### subplot(), subplots(), subplot_mosaic() 비교

||subplot()|subplots()|subplot_mosaic()|
|---|---|---|---|
|코드 방식|subplot을 하나씩 수동 지정|subplot을 배열로 일괄 생성|문자열/딕셔너리 기반으로 유연하게 레이아웃 구성|
|접근 방식|번호로 직접 접근(plt.subplot(1,2,1))|배열의 인덱싱(axes[0,1], axes[1], axes[0][1])|이름(label)으로 접근(ax['A'],ax['B'])|
|반환 값| 단일 Axes Subplot 객체|(Figure, Axes) 튜플(Axes 배열 또는 객체 배열)|(Figure, dict[str, Axes]) 형태로 반환|
|코드 구조|반복문 없이 단순한 코드지만 중복 많음|반복문 활용 가능. 배열 인덱싱 필요|레이아웃 정의가 직관적이고 가독성 좋음. 반복문도 가능|
|레이아웃 유연성|낮음|중간(격자형만 가능)|매우 높음(불균형, 중첩, 특정 영역 확대 등 자유롭게 설정 가능)|
|권장 사용|아주 단순한 1xN 또는 N행의 그래프|다수 subplot을 일정한 행렬 형태로 반복하여 그릴 때|복잡한 레이아웃, 대시보드 구성, 이름 기반으로 명확히 구분할 때|
|버전 요구사항|오래된 버전부터 사용 가능|오래된 버전부터 사용 가능|Matplotlib 3.4 이상|
|예시 코드|plt.subplot(2,1,1)|fig, axes = plt.subplots(2,2)|fig, axes = plt.subplot_mosaic([['A','B'],['C','B']]))|
|요약|subplot 하나씩 단순하게 붙이기|반복적인 플롯 그리기|이름으로 구역 구분, 복잡한 배치|

```python
nrow, ncol = 2,2
x = range(len(data2.index))
xlabels = list(data2.index)

fig, axes = plt.subplots(nrow, ncol, figsize=(15,8))
plt.suptitle('지역별 미세먼지량')
for i, col in enumerate(data2.columns):
    r = i//ncol
    c = i%ncol
    axes[r,c].plot(x, data2[col], label=col)
    axes[r,c].set_title(f'year={col}')
    axes[r,c].set_xlabel('area')
    axes[r,c].set_ylabel('micrometer')
    axes[r,c].set_xticks(x, xlabels, rotation=45, fontdict={'fontsize':8})
    axes[r,c].set_ylim(30,55)

plt.tight_layout()
plt.show()
```
